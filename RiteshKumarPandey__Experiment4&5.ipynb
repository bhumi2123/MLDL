**Train and Test Split**

```python
from sklearn.model_selection import train_test_split

def split_data(X, Y):
  X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, random_state=42, shuffle=True)
  print(" Training Set Feature Shape is {} \n Training Set Prediction Shape is {}".format(X_train.shape, Y_train.shape))
  print("\n Testing Set Feature Shape is {} \n Testing Set Prediction Shape is {}".format(X_test.shape, Y_test.shape))
  return X_train, X_test, Y_train, Y_test
```

**Classification**

```python
df = pd.read_csv('gdrive/MyDrive/Semester 7/data/Social_Network_Ads.csv')
df['Gender'] = df['Gender'].map({'Male':int(1) ,'Female':int(0)})
x = df[['Gender', 'Age', 'EstimatedSalary']]
y = df.Purchased

df
```

```python
X_train, X_test, Y_train, Y_test = split_data(x, y)
```

```python
#Single layer architecture
model = Sequential()
model.add(Dense(3, input_shape=(X_train.shape[1],)))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, Y_train, batch_size=32, epochs=100, verbose=1, validation_data=(X_test, Y_test))
loss, accuracy = model.evaluate(X_test,Y_test, verbose=0)
```

```python
print("Accuracy = "+str(accuracy*100.0)+"%")
```

```python
print('Layer 1:', model.layers[0].get_weights()[1])
print('Layer 2:', model.layers[1].get_weights()[1])
```

**Prediction**

```python
df = pd.read_csv('gdrive/MyDrive/Semester 7/data/spms.csv')
df = df.dropna(axis=0)
df = df[(df['year_of_graduation']==2018) | (df['year_of_graduation']==2019) | (df['year_of_graduation']==2020) | (df['year_of_graduation']==2021)]

x = df[['ga_score_fy', 'ga_score_sy', 'ga_score_ty']]
y = df.ga_score_graduation.values

X_train, X_test, Y_train, Y_test = split_data(x, y)
```

```python
# Creating model using the Sequential in tensorflow
def build_model_using_sequential():
  model = Sequential([
    Dense(3, kernel_initializer='normal', activation='relu'),
    Dropout(0.4),
    Dense(1, kernel_initializer='normal', activation='linear')
  ])
  return model

# build the model
ann_model = build_model_using_sequential()
```

```python
# loss function
msle = MeanSquaredLogarithmicError()
ann_model.compile(
    loss=msle, 
    optimizer=Adam(learning_rate=0.01), 
    metrics=[msle]
)

# train the ann_model
history = ann_model.fit(
    X_train.values, 
    Y_train, 
    epochs=100, 
    batch_size=128,
    validation_split=0.4
)
```

```python
ann_model.summary()
```

```python
y_predicted = ann_model.predict(X_test)

print('R2 Score:', metrics.r2_score(Y_test, y_predicted))
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, y_predicted))
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, y_predicted))
```
