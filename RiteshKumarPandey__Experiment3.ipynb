# **Performance Visualization**

```python
def visualization(result_df):
  fig = px.bar(result_df, 
             x='Model', 
             y='Accuracy')
  fig.update_layout(
      font_family="Times New Roman",
      title_font_family="Times New Roman",
      legend_title='',
      title = "Accuracy of Models"
  )
  fig.update_xaxes(title_font_family="Times New Roman", title='Models')
  fig.update_yaxes(title_font_family="Times New Roman", title='Accuracy')
                    

  fig.show()

  fig = px.bar(result_df, 
             x='Model', 
             y='F-Score')
  fig.update_layout(
      font_family="Times New Roman",
      title_font_family="Times New Roman",
      legend_title='',
      title = "F-Score of Models"
  )
  fig.update_xaxes(title_font_family="Times New Roman", title='Models')
  fig.update_yaxes(title_font_family="Times New Roman", title='F-Score')
                    

  fig.show()
```

# **Model Evaluation Function**

```python
from sklearn import metrics

def evaluate(name, y_actual, y_predicted):
    Model_name.append(name)
    Accuracy.append(100*metrics.accuracy_score(y_actual, y_predicted))
    F_score.append(100*metrics.f1_score(y_actual, y_predicted))
```

# **KNN**

```python
def knnmodel(X, Y):
  X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=None, shuffle=True)
  clf = KNeighborsClassifier(n_neighbors=2)
  clf.fit(X_train,Y_train)

  Y_pred = clf.predict(X)

  evaluate('KNN', Y, Y_pred)
```

# **Naive Byes**

```python
def naivebyesmodel(X, Y):
  X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=None, shuffle=True)
  clf = GaussianNB()
  clf.fit(X_train,Y_train)

  Y_pred = clf.predict(X)

  evaluate('Naive Byes', Y, Y_pred)
```

# **Random Forest Library**

```python
from sklearn.ensemble import RandomForestClassifier

def randomforestlibrary(X, Y):
  X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=None, shuffle=True)

  clf = RandomForestClassifier()
  clf.fit(X_train, Y_train)

  Y_pred = clf.predict(X)

  evaluate('Random Forest Library', Y, Y_pred)
```

# **Dataset 1** - Telecommunication Data

```python
df = pd.read_csv('gdrive/MyDrive/Semester 7/data/telcom.csv')
df['international plan'] = df['international plan'].map({'yes':1 ,'no':0})
df['voice mail plan'] = df['voice mail plan'].map({'yes':1 ,'no':0})

x = df[['international plan', 'voice mail plan', 'total day charge', 'total eve charge', 'total night charge', 'total intl charge']]
y = df.churn

x1 = df[['international plan', 'voice mail plan']]
x2 = df[['total day charge', 'total eve charge']]
x3 = df[['total night charge', 'total intl charge']]

df
```

```python
Model_name = []
Accuracy = []
F_score = []

naivebyesmodel(x, y)
knnmodel(x, y)

predicted_df = pd.DataFrame()
predicted_df_1 = pd.DataFrame()
predicted_df_2 = pd.DataFrame()

decisiontreemodel_30(x, y)
decisiontreemodel_30_0(x, y)
decisiontreemodel_30_42(x, y)
decisiontreemodel_40(x, y)
decisiontreemodel_40_0(x, y)
decisiontreemodel_40_42(x, y)
decisiontreemodel_35(x, y)
decisiontreemodel_(x1, y, '1')
decisiontreemodel_(x2, y, '2')
decisiontreemodel_(x3, y, '3')

predicted_df_2['count'] = predicted_df_2[['Decision Tree 1', 'Decision Tree 2', 'Decision Tree 3']].sum(axis=1)
predicted_df_2['final'] = predicted_df_2['count'].map({3:True, 2:True, 1:False, 0:False})
evaluate('Random Forest Attribute Split', y, predicted_df_2['final'])

predicted_df_1['count'] = predicted_df_1[['Decision Tree 30', 'Decision Tree 30_0', 'Decision Tree 30_42', 'Decision Tree 40', 'Decision Tree 40_0', 'Decision Tree 40_42', 'Decision Tree 35']].sum(axis=1)
predicted_df_1['final'] = predicted_df_1['count'].map({7: True, 6: True, 5:True, 4:True, 3:False, 2:False, 1:False, 0:False})
evaluate('Random Forest Tuple Split', y, predicted_df_1['final'])

predicted_df['count'] = predicted_df[['Decision Tree 30', 'Decision Tree 30_0', 'Decision Tree 30_42', 'Decision Tree 40', 'Decision Tree 40_0', 'Decision Tree 40_42', 'Decision Tree 35', 'Decision Tree 1', 'Decision Tree 2', 'Decision Tree 3']].sum(axis=1)
predicted_df['final'] = predicted_df['count'].map({10: True, 9: True, 8: True, 7: True, 6: True, 5:True, 4:False, 3:False, 2:False, 1:False, 0:False})
evaluate('Random Forest Combine', y, predicted_df['final'])

randomforestlibrary(x, y)

di = {'Model': Model_name, 'Accuracy': Accuracy, 'F-Score': F_score}
result_df = pd.DataFrame(di)
print(result_df)

visualization(result_df)
```
