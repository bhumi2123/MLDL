```python
import pandas as pd
import time
```

```python
df = pd.read_csv('gdrive/MyDrive/Semester 7/data/telcom.csv')
df
```

**Creating Features and Labels**

```python
x = df[['international plan', 'voice mail plan', 'total day charge', 'total eve charge', 'total night charge', 'total intl charge']]
x['international plan'] = x['international plan'].map({'yes':1 ,'no':0})
x['voice mail plan'] = x['voice mail plan'].map({'yes':1 ,'no':0})
x
```

```python
y = df.churn
y
```

**Training and Testing Data**

```python
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1) # 70% training and 30% test
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

**Decision Tree Classifier**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decision = DecisionTreeClassifier()
start_time = time.time()
decision = decision.fit(X_train,y_train)
elapsed_time = time.time() - start_time
print(f'{elapsed_time:.2f}s elapsed during training')
```

**Bagging Classifier**
Bagging is an Ensemble Learning technique which aims to reduce the error learning through the implementation of a set of homogeneous machine learning algorithms. The key idea of bagging is the use of multiple base learners which are trained separately with a random sample from the training set, which through a voting or averaging approach, produce a more stable and accurate model.

The bagging process is quite easy to understand, first it is extracted “n” subsets from the training set, then these subsets are used to train “n” base learners of the same type. For making a prediction, each one of the “n” learners are feed with the test sample, the output of each learner is averaged (in case of regression) or voted (in case of classification).

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bagging = BaggingClassifier(base_estimator=decision)
start_time = time.time()
bagging.fit(X_train, y_train)
elapsed_time = time.time() - start_time
print(f'{elapsed_time:.2f}s elapsed during training')
```

```python
bagging_bootstrap = BaggingClassifier(base_estimator=decision, bootstrap=True)
start_time = time.time()
bagging_bootstrap.fit(X_train, y_train)
elapsed_time = time.time() - start_time
print(f'{elapsed_time:.2f}s elapsed during training')
```

**Boosting**
Boosting is an Ensemble Learning technique that, like bagging, makes use of a set of base learners to improve the stability and effectiveness of a ML model. The idea behind a boosting architecture is the generation of sequential hypotheses, where each hypothesis tries to improve or correct the mistakes made in the previous one.

The key differentiator between boosting-based techniques is the way in which errors are penalized (by modifying weights or minimizing a loss function) as well as how the data is sampled.

**Ada Boost Classifier**
AdaBoost implements a vector of weights to penalize those samples that were incorrectly inferred (by increasing the weight) and reward those that were correctly inferred (by decreasing the weight).

Each base learner in the sequence will have assigned a weight, the higher the performance, the higher the weight and the greater the impact of this base learner for the final decision.

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
adaboost = AdaBoostClassifier(base_estimator=decision)
start_time = time.time()
adaboost.fit(X_train, y_train)
elapsed_time = time.time() - start_time
print(f'{elapsed_time:.2f}s elapsed during training')
```

**Gradient Boost Classifier**
The Gradient Boosting method does not implement a vector of weights like AdaBoost does. As its name implies, it implements the calculation of the gradient for the optimization of a given loss function. The core idea of Gradient Boosting is based on minimizing the residuals of each learner base in a sequential way, this minimization is carried out through the calculation of the gradient applied to a specific loss function (either for classification or regression). Then each base learner added to the sequence will minimize the residuals determined by the previous base learner. This will be repeated until the error function is as close to zero or until a specified number of base learners is completed.

Gradient Boosting works with decision trees by default, that is why in the implementation we do not define a specific base learner.

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gradient = GradientBoostingClassifier()
start_time = time.time()
gradient.fit(X_train, y_train)
elapsed_time = time.time() - start_time
print(f'{elapsed_time:.2f}s elapsed during training')
```

**XGB Boost Classifier**
XgBoost stands for Extreme Gradient Boosting.
XGBoost is an implementation of Gradient Boosted decision trees.In this algorithm, decision trees are created in sequential form. Weights play an important role in XGBoost. Weights are assigned to all the independent variables which are then fed into the decision tree which predicts results. The weight of variables predicted wrong by the tree is increased and these variables are then fed to the second decision tree. These individual classifiers/predictors then ensemble to give a strong and more precise model.

```python
from xgboost import XGBClassifier
```

```python
xgb = XGBClassifier(n_estimators=1000, max_depth=15, eta=0.4, subsample=0.9, colsample_bytree=0.8)
start_time = time.time()
xgb.fit(X_train, y_train)
elapsed_time = time.time() - start_time
print(f'{elapsed_time:.2f}s elapsed during training')
```

**Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
randomforest = RandomForestClassifier()
start_time = time.time()
randomforest.fit(X_train, y_train)
elapsed_time = time.time() - start_time
print(f'{elapsed_time:.2f}s elapsed during training')
```

**Performance Evaluation**

```python
Model_name = []
Accuracy = []
AUC = []
F_score = []
```

```python
def evaluate(model, title):
  start_time = time.time()
  y_predict = model.predict(X_test)
  elapsed_time = time.time() - start_time
  print(title)
  print('Accuracy: ', metrics.accuracy_score(y_test, y_predict))
  print('AUC: ', metrics.auc(y_test, y_predict))
  print('F-Score: ', metrics.f1_score(y_test, y_predict))
  print(f'{elapsed_time:.2f}s elapsed during testing')
  print('\n')
  Model_name.append(title)
  Accuracy.append(100*metrics.accuracy_score(y_test, y_predict))
  AUC.append(metrics.auc(y_test, y_predict))
  F_score.append(100*metrics.f1_score(y_test, y_predict))
```

```python
evaluate(decision, 'Decision Tree Classifier')
evaluate(bagging, 'Bagging Classifier')
evaluate(bagging_bootstrap, 'Bagging with Bootstrap')
evaluate(adaboost, 'Ada Boost Classifier')
evaluate(gradient, 'Gradient Boost Classifier')
evaluate(xgb, 'XGB Boost Classifier')
evaluate(randomforest, 'Random Forest Classifier')
```

**Performance Visualization**

```python
di = {'Model': Model_name, 'Accuracy': Accuracy, 'AUC': AUC, 'F-Score': F_score}
result_df = pd.DataFrame(di)
result_df
```
